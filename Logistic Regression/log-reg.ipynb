```python
dataset.drop("Cabin", axis=1 , inplace=True)
```

```python
dataset.head(10)
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.isnull().sum()
```

```python

#dummies
sex = pd.get_dummies(dataset["Sex"], drop_first=True)
sex.head(10)
```

```python
embark = pd.get_dummies(dataset["Embarked"] , drop_first=True)
embark.head(10)
```

```python
pclass = pd.get_dummies(dataset["Pclass"] , drop_first=True)
pclass.head(10)
```

```python
#concatinate all these rows into dataset

dataset = pd.concat([dataset , sex , embark , pclass] , axis=1)
dataset.head(10)
```

```python
dataset.drop(["PassengerId" , "Pclass", "Name" , "Sex", "Ticket" , "Embarked" ] , axis=1 , inplace=True)
```

```python
dataset.head(10)
```

##Train Data

```python
y = dataset["Survived"]
x= dataset.drop("Survived",axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_tr , x_tst , y_tr , y_tst = train_test_split(x ,y , test_size=0.25 , random_state=42)
```

```python

x_tr.columns = x_tr.columns.astype(str)

```

```python
model = LogisticRegression(max_iter=200)
```

```python
model.fit(x_tr , y_tr)
```

```python
x_tst.columns = x_tst.columns.astype(str)
```

```python
predict = model.predict(x_tst)
```

```python
from sklearn.metrics import classification_report , confusion_matrix
```

```python
confusion_matrix(y_tst , predict)
```
